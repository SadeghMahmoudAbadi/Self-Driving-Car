```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
# get the same random number each run
np.random.seed(0)
```

```python
# number of points
n_pts = 500
# declare X values
X = np.linspace(-3, 3, n_pts)
# declare Y values as sin(X) plus a noise
Y = np.sin(X) + np.random.uniform(-0.5, 0.5, n_pts)
# plot the points
plt.scatter(X, Y)
```

```python
# declare the model
model = Sequential()
# add the first layer with 50 neurons, activation function sigmoid, and input dimension 1 for X
model.add(Dense(50, activation='sigmoid', input_dim=1))
# add the second layer with 30 neurons and activation function sigmoid
model.add(Dense(30, activation='sigmoid'))
# add the output layer for Y
model.add(Dense(1))
```

```python
# compile the model using Adam optimizer with learnin rate 0.01 and loss function MSE, mean square error
model.compile(Adam(0.01), loss='mse')
```

```python
# fit the model to the samples in 50 epochs
model.fit(X, Y, epochs=50)
```

```python
# display the sample points
plt.scatter(X, Y)
# plot the prediction points in red circles
plt.plot(X, model.predict(X), 'ro')
```
